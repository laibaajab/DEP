```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes= True)
```

```python
weather = pd.read_csv("weatherHistory.csv")
```

```python
weather.head()
```

```python
weather.info()
```

```python
top_summaries = weather['Summary'].value_counts().index[:10]
filtered_weather = weather[weather['Summary'].isin(top_summaries)]
sns.barplot(x=filtered_weather['Summary'], y=filtered_weather['Temperature (C)'])
plt.xticks(rotation=90)
plt.show()

```

```python
sns.displot(weather['Wind Bearing (degrees)'])
plt.show()
```

```python
sns.jointplot(x=weather['Visibility (km)'], y=weather['Temperature (C)'])
plt.show()
```

```python
sns.jointplot(x=weather['Humidity'], y=weather['Temperature (C)'] ,kind = "hex")
plt.show()
```

```python
sns.pairplot(weather[['Humidity' , 'Temperature (C)', 'Wind Speed (km/h)']])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
top_summaries = weather['Summary'].value_counts().index[:10]
filtered_weather = weather[weather['Summary'].isin(top_summaries)]
sns.stripplot(x='Summary', y='Temperature (C)', data=filtered_weather, hue='Summary', dodge=True, palette='tab10')
plt.xticks(rotation=90)
plt.show()

```

```python
sns.stripplot(x=weather['Humidity'], y=weather['Pressure (millibars)'], jitter= True)
plt.show()
```

```python
sns.boxplot(x='Humidity', y='Temperature (C)', data=weather ,hue='Precip Type')
plt.show()
```

```python
sns.barplot(x='Humidity', y='Temperature (C)', data=weather ,hue='Precip Type')
plt.show()
```

```python
sns.countplot(x='Precip Type', data= weather)
plt.show()
```

```python
sns.pointplot(x='Humidity', y='Temperature (C)', data=weather, hue='Precip Type')
plt.show()
```

```python
sns.lmplot(x='Humidity', y='Temperature (C)', data=weather)
plt.show()
```

```python
sns.lmplot(x='Humidity', y='Apparent Temperature (C)', data=weather, hue='Loud Cover')
plt.show()
```
